```python
import matplotlib.pyplot as plt
values=[20,55,31,90]
labels=['c','java','c++','pyhton']
plt.pie(values,labels=labels,shadow=True)
print(plt.show())
```

```python
import matplotlib.pyplot as plt
values=[20,55,31,90]
labels=['c','java','c++','pyhton']
explode=(0,0.3,0,0)
plt.pie(values,labels=labels,shadow=True,explode=explode)
print(plt.show())
```

```python
#adding legend
import matplotlib.pyplot as plt
values=[20,55,31,90]
labels=['c','java','c++','pyhton']
explode=(0,0.3,0,0)
plt.pie(values,labels=labels,shadow=True,explode=explode)
plt.legend(loc='best')
print(plt.show())
```
