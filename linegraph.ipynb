```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,4,6,8,10]
plt.plot(x,y,'bo')
plt.axis([0,5,0,10])
print(plt.show)
```

```python
#adding title  labels for axis
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,4,6,8,10]
plt.plot(x,y,'bo')
plt.title('Line Graph')
plt.axis([0,5,0,10])
plt.xlabel('x label values')
plt.ylabel('y label values')
print(plt.show)
```

```python
#multiline value graph
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,4,6,8,10]
x1=[1,3,5,7,8]
y1=[1,4,7,9,10]
plt.plot(x,y,'b-')
plt.plot(x1,y1,'r-')
plt.title('Line Graph')
plt.axis([0,10,0,10])
plt.xlabel('x label values')
plt.ylabel('y label values')
print(plt.show)
```

```python
#line width for line plotted
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,4,6,8,10]
plt.plot(x,y,'b-',linewidth=4)
plt.title('Line Graph')
plt.axis([0,5,0,10])
plt.xlabel('x label values')
plt.ylabel('y label values')
print(plt.show)
```
