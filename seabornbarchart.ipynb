```python
#barplot seaborn
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
days=['sunday','monday','tuesday','wensday','thursday','friday','satday']
tips=[300,302,405,390,357,324,384]
tips=pd.DataFrame({'days':days,'tips':tips},columns=['days','tips'])
print(tips)
```

```python
sns.set(style="whitegrid")
ax=sns.barplot(x='days',y='tips',data=tips)
```

```python
ax=sns.barplot(x='tips',y='days',data=tips,capsize=.2,palette="Blues_d")
```

```python
ax=sns.barplot(x='tips',y='days',data=tips,capsize=.2)
```

```python
from numpy import median
ax=sns.barplot(x='days',y='tips',capsize=0.15,estimator=median,data=tips,order=['monday','tuesday','satday','wensday','friday','thursday','sunday'])
```

```python
days=['sunday','monday','tuesday','wensday','thursday','friday','satday']
tips=[300,302,405,390,357,324,384]
holiday=['no','yes','no','yes','no','yes','no']
tips=pd.DataFrame({'days':days,'tips':tips,'holidays':holiday},columns=['days','tips','holidays'])
print(tips)
```

```python
ax=sns.barplot(x='days',y='tips',capsize=0.15,estimator=median,hue='holidays',data=tips,order=['monday','tuesday','satday','wensday','friday','thursday','sunday'])
```
