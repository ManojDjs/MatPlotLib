```python
#fillplot implementation
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,4,6,8,10]
plt.plot(x,y,'b-')
plt.ylim(0,10)
plt.fill_between(x,y,color='red',alpha='0.2')
print(plt.show())
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,4,6,8,10]
y2=[3,5,6,7,10]
plt.plot(x,y,color='red')
plt.plot(x,y2,color='blue')
plt.ylim(0,10)
plt.fill_between(x,y,y2,color='red',alpha='0.2')
print(plt.show)
```
