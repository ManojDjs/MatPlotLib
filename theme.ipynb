```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
print(plt.style.available)
x=[1,2,3,4,5]
y=[x**2 for x in x]
plt.plot(x,y,'r-')
#for gird implementation use grid method
plt.grid(True)
print(plt.show())
```
