```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[x**2 for x in x]
plt.plot(x,y,'r-')
#for gird implementation use grid method
plt.grid(True)
print(plt.show())
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[x**2 for x in x]
plt.plot(x,y,'r-',label='student')
plt.legend(loc='best')
#for gird implementation use grid method
plt.grid(True)
print(plt.show())
```

```python
#for color of grid and major 
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[x**2 for x in x]
plt.plot(x,y,'r-',label='student')
plt.legend(loc='best')
#for gird implementation use grid method
plt.grid(which='major',color='blue')
print(plt.show())
```

```python
#to also view the minor lines
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[x**2 for x in x]
plt.plot(x,y,'r-',label='student')
plt.legend(loc='best')
#for gird implementation use grid method
plt.grid(which='major',color='blue')
plt.minorticks_on()
plt.grid(which='minor',color='green',linestyle='--')
print(plt.show())
```
